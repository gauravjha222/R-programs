**Pie chart of number of forward gears in automobiles**

Graph conclusion: 3 gear automobile present higher quantity in over all automobiles.

```r
count<-table(mtcars$gear)
pie(count, labels = paste0(count, "%"))

legend("topleft", legend = c("3 gears", "4 gears", "5 gears"),
       fill =  c("white", "lightblue", "mistyrose"))
```

**Histogram plot of mpg(miles per gallon) and frequency of automobile**

Conclusion: Most of the automobile mpg have in range between 15-20

```r
hist(mtcars$mpg,
     col='steelblue',
     main='Histogram',
     xlab='mpg',
     ylab='Frequency')
```

**quantiles of mtcars dataset**
* It divide the sample in equal group or size.
* In the quantiles, the 25th percentile is called as lower quartile
* 50th percentile is called as Median
* 75th Percentile is called as the upper quartile.

```r
quantile(mtcars$wt)
```

**graph for wt using boxplot**

* The box in the Box Plot extends from the lower quartile to the upper quartile.
* The lower quartile is the point where one-quarter of the values are below it.
* The difference between the upper quartile and the lower quartile is called the inter-quartile range.

```r
boxplot(mtcars$wt,
        main='Distribution of wt values',
        ylab='wt',
        col='steelblue',
        border='black')
```
