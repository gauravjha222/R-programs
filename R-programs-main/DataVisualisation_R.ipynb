# **PIE CHARTS**

```r
#convert cyl to factor
library(dplyr)

mtcars <- mtcars %>%
  mutate(cyl_factor = as.factor(cyl))
ggplot(data=mtcars,aes(x=cyl_factor,fill=cyl_factor))+geom_bar(position="dodge")
```

**stack bar charts**

```r
ggplot(data=mtcars,aes(x=" ",fill=cyl_factor))+geom_bar(position="stack")
```

**Converting a bar chart to pie chart**

```r
ggplot(data=mtcars,aes(x=" " ,fill=cyl_factor))+geom_bar(position="stack")+coord_polar(theta="y")
```

**Customizing pie charts with themes**

```r
ggplot(data=mtcars,aes(x=" " ,fill=cyl_factor))+geom_bar(position="stack")+coord_polar(theta="y")+theme_minimal()
```

```r
ggplot(data=mtcars,aes(x=" " ,fill=cyl_factor))+geom_bar(position="stack")+coord_polar(theta="y") + scale_fill_brewer(palette="Dark2")
```

# **SCATTER PLOTS**

```r
qplot(mpg,wt,data=mtcars)
```

**changing the shape of points**

```r
ggplot(mtcars,aes(x=mpg,y=wt))+geom_point(shape=19,color="blue")
```

**the third variable as a shape**

```r
mtcars$cyl_factor<-factor(mtcars$cyl)
ggplot(mtcars,aes(x=mpg,y=wt,shape=cyl_factor))+geom_point()
```

**the third variable as a color**

```r
ggplot(mtcars,aes(x=mpg,y=wt,color=cyl))+geom_point(shape=19)
```

```r
ggplot(mtcars,aes(x=mpg,y=wt,color=cyl_factor))+geom_point(shape=19)
```

**labels and title function in scatter plot**

```r
ggplot(mtcars,aes(x=mpg,y=wt,color=cyl_factor))+geom_point(shape=19)+labs(color="Cylinder")+xlab("MIles per gallon")+ylab("weight")+ggtitle("Scatterplot")
```

# **CUSTOMIZE PLOTS**

```r
install.packages("ggplot2")
```

```r
library(ggplot2)
```

```r
ggplot(mtcars,aes(x=wt,y=mpg))+
  geom_point(aes(color=factor(cyl)))+
  labs(
  x="weight (1000 lbs)",
  y="miles/(US) gallon",
  color="Cylinders",
  title="mileage by weight and cylinders",
  subtitle="source : 1974 motors trend US magazine"
  )
```

**text labels: geom_text()**

```r
ggplot(mtcars,aes(x=wt,y=mpg))+
  geom_point(aes(color=factor(cyl)))+
  geom_text(aes(label=rownames(mtcars)))
```

**To remove overlapping : check_overlap=TRUE**

```r
ggplot(mtcars,aes(x=wt,y=mpg))+
  geom_point(aes(color=factor(cyl)))+
  geom_text(aes(label=rownames(mtcars)),
check_overlap=TRUE,hjust="inward")
```

**Custom annotations Functions**
*   geom_vline(xintercept)
*   geom_hline(yintercept)

*   geom_abline(slope,intercept)
*   geom_rect()

*   annotate(geom)







```r
#example
ggplot(mtcars,aes(x=mpg))+
  geom_histogram(bins=10)+
  labs(x="miles/gallon(US)",
y="count",
title="Distribution of miles/gallon")+
  geom_vline(aes(xintercept=19.2),color="red")+
  annotate(geom="text",label="median=19.2",x=21,y=7,hjust=0,color="red")
```
