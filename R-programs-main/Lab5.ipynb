**Create a data frame with the fields of weight , height and sex then create a plot of weight on the x-axis and height on the y-axis . use different point character or color to distinguish between males and females and provide a matching legend. Label the axes and give the plot a title**

```r
df <- data.frame(
  weight = c(50, 60, 70, 80, 60),
  height = c(170, 161, 168, 145, 140),
  sex = c("Male", "Female", "Female", "Male", "Female"),
  stringsAsFactors = FALSE
)

plot(df$weight, df$height, col = ifelse(df$sex == "Male", "blue", "red"), pch = ifelse(df$sex == "Male", 16, 17), 
     xlab = "Weight", ylab = "Height", main = "Weight vs Height")

legend("topright", legend = c("Male", "Female"), col = c("blue", "red"), pch = c(16, 17))
```

**Create a plot using ggplot2 for the same question above that consist of weight on the x-axis and height on the y-axis. use different point character or color to distinguish between males and females and provide a matching legend. Label the axes and give a plot a title**

```r
library(ggplot2)

df <- data.frame(
  weight = c(50, 60, 70, 80, 60),
  height = c(170, 161, 168, 145, 140),
  sex = c("Male", "Female", "Female", "Male", "Female"),
  stringsAsFactors = FALSE
)

ggplot(df, aes(x = weight, y = height, color = sex, shape = sex)) +
  geom_point(size = 3) +
  scale_color_manual(values = c("blue", "red")) +
  scale_shape_manual(values = c(16, 17)) +
  labs(x = "Weight", y = "Height", title = "Weight vs Height") +
  theme_minimal()
```

**Write a R code that will plot education on the x-axis and income on y-axis , with both x-axis and y-axis limits fixed to be [0,100]. Provide appropriate axis labels for jobs with a prestige value of less than or equal to 80 , use a black '*' as the point character. for jobs with prestige greater than 80, use a blue '@'**

```r
education <- c(60, 80, 70, 90, 50)
income <- c(40, 60, 70, 80, 30)
prestige <- c(70, 85, 60, 95, 75)
plot(education, income, xlim = c(0, 100), ylim = c(0, 100), xlab = "Education", ylab = "Income", main = "Education vs Income")

points(education[prestige <= 80], income[prestige <= 80], pch = "*", col = "black")
points(education[prestige > 80], income[prestige > 80], pch = "@", col = "blue")
```
